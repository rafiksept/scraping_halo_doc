```python
import pandas as pd
```

```python
df = pd.read_csv("data_nama_obat3.csv")
```

```python
df
```

```python
def replace_symbols_with_space(value):
    if isinstance(value, str):
        return ''.join(char if char.isalnum() else ' ' for char in value)
    else:
        return value

# Applying the function to each element in the DataFrame
df = df.applymap(replace_symbols_with_space)

# Displaying the modified DataFrame
print(df)
```

```python
df
```

```python
def replace_space_with_dash(value):
    if isinstance(value, str):
        return value.replace(' ', '-')
    else:
        return value

# Applying the function to each element in the DataFrame
df = df.applymap(replace_space_with_dash)

# Displaying the modified DataFrame
df
```

```python
df.to_csv("name_obat_fix.csv")
```
